```python
import requests
from bs4 import BeautifulSoup
```

```python
re = requests.get("https://www.javatpoint.com/java-string")
```

```python
print(re.text)
```

```python
soup = BeautifulSoup(re.text,"html.parser")
```

```python
text = soup.find_all("table",class_="alt")
```

```python
for values in soup.find_all("table",class_="alt"):
    print(values.prettify())
```

```python
fd = open("maining.html","w")
fd.write(str(text))
fd.close()
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

### comapring tips vs total_bill

```python
plt.figure(figsize=(14,12))
sns.barplot(x=tips.tip,y=tips.total_bill,width=0.8,alpha=0.4)
plt.xticks(fontsize=6, rotation=60, weight='bold')
plt.show()
```

### the inference is that the person with highest amount of tips have a high total_bill

```python
categorising_smoker=tips["smoker"].value_counts()
```

```python
categorising_smoker
```

#### checking the total amount,tips given by perosn on basis of their smoking habits

```python
sns.relplot(x=tips.tip,y=tips.total_bill,data=tips,hue="smoker",edgecolor="black")
plt.savefig("relplot.png")
```

### Tips based on no of people who came to dinner

```python
sns.relplot(x=tips.tip,y=tips.total_bill,data=tips,style="size",edgecolor="black",color="y")
```

```python
sns.relplot(x=tips.tip,y=tips.total_bill,data=tips,style="size",hue="time",edgecolor="black",color="y")
```

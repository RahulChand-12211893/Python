## Filling the values where it is null

#### As for Categorical data we can fill the data with a value if we have any, if not we can prefer the mode.
#### Mean imputation is often used when the missing values are numerical and the distribution of the variable is approximately normal.
#### Median imputation is preferred when the distribution is skewed, as the median is less sensitive to outliers than the mean

```python
# now in our dataset we have numerical values missing so we find mean_data
mean_data = titanic_data["Age"].mean()
print(mean_data)
```

```python
## filling the mean data wherever it is missing in the age columns
titanic_data["Age"].fillna(mean_data,inplace=True)
```

```python
titanic_data.head()
```

```python
# now verifying the data
titanic_data.isnull().sum()
```

#### now for categorical data missing

```python
titanic_data.Cabin
```

#### now we can either go for mode or drop data i prefer dropping the unecessary values but if we drop then ther will be alot of data loss to prevent that we can use subset

```python
titanic_data.dropna(how="any",inplace=True)
# how = any it drops all the rows where any  column values are missing
# how= all it drops the rows where all  the speciifed column values is missing
```

# Data Manipulation

```python
titanic_data.Fare = titanic_data.Fare.str.rstrip("$").astype(float)
```

```python
titanic_data.Fare
```

#### now we try to normalize the values

```python
max_fare = titanic_data.Fare.max()
min_fare = titanic_data.Fare.min()
def normalize(x):
    return (x-min_fare)/(max_fare-min_fare)
```

## Adding of columns

```python
titanic_data["normalized_price"] = titanic_data.Fare.apply(normalize)
titanic_data
```

```python
# Sample DataFrame
data = {'A': [1, 2, 3, 1, 2], 'B': [4, 5, 6, 4, 5]}
df = pd.DataFrame(data)
 # Define a function to calculate the average of a row
def average_row(row):
    return row.mean()
 # Apply the function row-wise.
df['Row_Average'] = df.apply(average_row, axis=1) ## for axis =1 columns
df
```

.map usefiul for transforming one column based on values from another

```python
mapping_dict = {1: 'one', 2: 'two', 3: 'three'}
df['Encoded_A'] = df['A'].map(mapping_dict)
df
```

.applymap for entire dataframe
 When you want to apply a function to each element in the entire DataFrame, you can use
 .applymap().

```python
data = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(data)
# Define a function to add 10 to a value
def add_10(x):
    return x + 10
# Apply the function to the entire DataFrame
df = df.applymap(add_10)
df
```

# Combining DataFrames

```python
import pandas as pd
data1 = {"A":[1,2,3],"B":[4,5,6],"C":[7,8,9]}
data2 = {"A":[10,11,12],"B":[13,14,15],"D":[16,17,18]}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

horizontal_concat = pd.concat([df1,df2],axis=1) # axis = 0 for rows and axis = 1 for columns
horizontal_concat
```

### to get proper index

```python
horizontal_concat.reset_index(drop=True)
```

### Merger using inner join

```python
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [2, 3, 4], 'Age': [25, 30, 22]})
print("Inner Join")
result = pd.merge(df1,df2,on="ID",how="inner")
```

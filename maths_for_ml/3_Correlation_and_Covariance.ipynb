## Correlation 


#### Pearson Correlation Coefficient

```python
df.corr(method="pearson",numeric_only=True) ## diagonal elemensta lways 1
```

#### Spearman Correlation Coefficient

```python
corr = df.corr(method="spearman",numeric_only=True)
```

```python
corr
```

```python
corr1 = df[["Year","Spending_USD"]].corr()
```

```python
sns.histplot(df["Year"],kde=True,bins=10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(corr1,annot=True,xticklabels={"Year","Spending_USD"},vmin=0.5)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(corr,annot=True,xticklabels={"Year","Spent","Lifeline"},vmin=0.5)
plt.show()
```

```python
plt.figure(figsize=(10,5))
ax = sns.heatmap(corr, 
            cmap='coolwarm',  # Colormap
            annot=True,       # Add annotations
            fmt=".2f",        # Format of annotations
            linewidths=0.5,   # Line width between cells
            linecolor='black',# Line color between cells
            cbar_kws={'label': 'Intensity'})
ax.xaxis.tick_top()
ax
```

```python
import logging
```

#### Five levels of logging
##### DEBUG: detailed information , typically of interest only when diagonising problems
##### INFO: Confirmation that things are working as expected
##### WARNING: an indication that something unexpected happended or inidicative of some problem in the near future
##### ERROR: Due to a more serious problem , the software has not been able to perform some function
##### CRITICAL: A serious error ,indicating that the program itself may be unable to continue running


### Default logging level is warning

```python
logging.basicConfig(filename='just_learning_logging.log',level=logging.DEBUG,format='%(asctime)s:%(levelname)s:%(message)s')
#the asctime provide time information , the levelname provied the level which we are dealing and the message just prints the message ,the s here is used to tell logger to
#everything as a string
```

```python
def add(num1,num2):
    return num1+num2
def sub(num1,num2):
    return num1-num2
def mul(num1,num2):
    return num1*num2
def div(num1,num2):
    return num1/num2
def mod(num1,num2):
    return num1%num2

num1=20
num2=5
add_result=add(num1,num2)
logging.debug('add {}+{}={}'.format(num1,num2,add_result))
sub_result=sub(num1,num2)
logging.debug('sub {}-{}={}'.format(num1,num2,sub_result))
mul_result=mul(num1,num2)
logging.debug('mul {}*{}={}'.format(num1,num2,mul_result))
div_result=div(num1,num2)
logging.debug('div {}/{}={}'.format(num1,num2,div_result))
mod_result=mod(num1,num2)
logging.debug('mod {}+{}={}'.format(num1,num2,mod_result))
```
